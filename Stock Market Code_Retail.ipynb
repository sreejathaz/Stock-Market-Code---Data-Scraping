```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
%matplotlib inline
```

```python
# Create start and end variables
start= datetime.datetime(2019,1,1)
end = datetime.datetime(2020,12,31)
```

# Gather Stock Information

```python
#STOCK INFORMATION ON SELECT FIRMS IN THE RETAIL INDUSTRY BASED ON MARKET CAP

# Home Depot
HD = data.DataReader('HD','yahoo',start,end)
# Walmart
WMT = data.DataReader('WMT','yahoo',start,end)
# Dollar Tree
DLTR = data.DataReader('DLTR','yahoo', start,end)
# Wallgreens
WBA = data.DataReader('WBA', 'yahoo', start,end)
# Costco
COST = data.DataReader('COST','yahoo',start,end)
# Nordstrom
JWN = data.DataReader('JWN','yahoo', start,end)
# Ross
ROST = data.DataReader('ROST','yahoo', start,end)
# Amazon
AMZN = data.DataReader('AMZN','yahoo', start,end)
```

```python
tickers = ['HD','WMT','DLTR','WBA','COST','JWN', 'ROST','AMZN'] 
```

```python
consumer_stocks = pd.concat([HD,WMT,DLTR,WBA,COST,JWN, ROST, AMZN], axis = 1, keys = tickers)
```

```python
consumer_stocks.columns.names = ['Consumer Ticker','Stock Info']
```

```python
consumer_stocks.head()
```

```python
consumer_stocks.to_csv('Consumer_Stock_Retail.csv')
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Optional Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
for tick in tickers:
    consumer_stocks[tick]['Adj Close'].plot(label = tick, figsize = (12,5))
plt.legend()
```

```python
for tick in tickers:
    consumer_stocks[tick]['Volume'].plot(label = tick, figsize = (12,4))
plt.legend()
```

```python

# Create heatmap
sns.heatmap(consumer_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').corr(),annot = True,lw = 1)
```

```python

# Create cluter map
sns.clustermap(consumer_stocks.xs(key = 'Close', axis = 1, level = 'Stock Info').corr(), annot = True, lw = 1)
```
